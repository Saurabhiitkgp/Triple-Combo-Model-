# Features Importance for DTSM

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split

y = df.iloc[:, 2:3]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
# Train a random forest regressor to determine feature importance
rf = RandomForestRegressor(random_state=5)
rf.fit(X_train, y_train)
# Get feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame with feature names and importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the features by importance in descending order
feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)

# Print the feature importance ranking
print("Feature Importance Ranking:")
print(feature_importance_df)

# Select the top k features based on importance
# k = 3  # Select the top 3 features
# selected_features = feature_importance_df.head(k)['Feature'].tolist()

# # Select the top k features from the original dataset
# X_selected = X[selected_features]

# # Print the selected features
# print(f"\nSelected Features (Top {k}):")
# print(X_selected.head())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

#X = df[[ 'DNPH', 'CNTC', 'SPAR', 'CFTC', 'KSDR']]
X = df.iloc[:,3:]
y = df.iloc[:, 1:2]
# Define the hyperparameters to test
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Create a random forest regression model
rf_model = RandomForestRegressor(random_state=11)

# Create the grid search cross-validation object
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=8, n_jobs=-1)

# Fit the grid search cross-validation object to the training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the performance of the best model on the training and testing data
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

train_mae = metrics.mean_absolute_error(y_train, y_train_pred)
test_mae = metrics.mean_absolute_error(y_test, y_test_pred)
train_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
train_r2 = metrics.r2_score(y_train, y_train_pred)
test_r2 = metrics.r2_score(y_test, y_test_pred)

print('Training MAE: ', train_mae)
print('Testing MAE: ', test_mae)
print('Training RMSE: ', train_rmse)
print('Testing RMSE: ', test_rmse)
print('Training R-squared score: ', train_r2)
print('Testing R-squared score: ', test_r2)
```

```python
y_pred_all = best_model.predict(X)
stacked_values = np.column_stack((y.squeeze(), y_pred_all))
print("Original y   Estimated y   Error Percentage")
for row in stacked_values:
    error_percentage = abs(row[0] - row[1]) / row[0] * 100
    print("{:<12} {:<12} {:.2f}%".format(row[0], row[1], error_percentage))
```

# For DTCO

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split

X = df[[ 'DNPH', 'CNTC', 'SPAR', 'CFTC', 'KSDR']]
#X = df.iloc[:,3:]
y = df.iloc[:, 1:2]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
# Define the hyperparameters to test
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Create a random forest regression model
rf_model = RandomForestRegressor(random_state=11)

# Create the grid search cross-validation object
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=8, n_jobs=-1)

# Fit the grid search cross-validation object to the training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the performance of the best model on the training and testing data
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

train_mae = metrics.mean_absolute_error(y_train, y_train_pred)
test_mae = metrics.mean_absolute_error(y_test, y_test_pred)
train_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
train_r2 = metrics.r2_score(y_train, y_train_pred)
test_r2 = metrics.r2_score(y_test, y_test_pred)

print('Training MAE: ', train_mae)
print('Testing MAE: ', test_mae)
print('Training RMSE: ', train_rmse)
print('Testing RMSE: ', test_rmse)
print('Training R-squared score: ', train_r2)
print('Testing R-squared score: ', test_r2)

# Print the name of the best model
print('Best model:', type(best_model).__name__)
```

```python
y_pred_all = best_model.predict(X)
stacked_values = np.column_stack((y.squeeze(), y_pred_all))
print("Original y   Estimated y   Error Percentage")
for row in stacked_values:
    error_percentage = abs(row[0] - row[1]) / row[0] * 100
    print("{:<12} {:<12} {:.2f}%".format(row[0], row[1], error_percentage))
```

```python
import pandas as pd

# Create a DataFrame from the stacked values
df_result = pd.DataFrame(stacked_values, columns=["Original y", "Estimated y"])
df_result["Error Percentage"] = df_result.apply(
    lambda row: abs(row["Original y"] - row["Estimated y"]) / row["Original y"] * 100,
    axis=1,
)

# Save the DataFrame to a CSV file
df_result.to_csv("DTCO.csv", index=False)
```

# For DTSM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split

X = df[[ 'SPAR', 'CNTC', 'KSDR', 'DNPH', 'CFTC']]
y = df.iloc[:, 2:3] # DTSM
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
# Define the hyperparameters to test
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Create a random forest regression model
rf_model = RandomForestRegressor(random_state=11)

# Create the grid search cross-validation object
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=8, n_jobs=-1)

# Fit the grid search cross-validation object to the training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the performance of the best model on the training and testing data
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

train_mae = metrics.mean_absolute_error(y_train, y_train_pred)
test_mae = metrics.mean_absolute_error(y_test, y_test_pred)
train_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
train_r2 = metrics.r2_score(y_train, y_train_pred)
test_r2 = metrics.r2_score(y_test, y_test_pred)

print('Training MAE: ', train_mae)
print('Testing MAE: ', test_mae)
print('Training RMSE: ', train_rmse)
print('Testing RMSE: ', test_rmse)
print('Training R-squared score: ', train_r2)
print('Testing R-squared score: ', test_r2)

# Print the name of the best model
print('Best model:', type(best_model).__name__)
```

```python
y_pred_all = best_model.predict(X)
stacked_values = np.column_stack((y.squeeze(), y_pred_all))
print("Original y   Estimated y   Error Percentage")
for row in stacked_values:
    error_percentage = abs(row[0] - row[1]) / row[0] * 100
    print("{:<12} {:<12} {:.2f}%".format(row[0], row[1], error_percentage))
```

```python
import pandas as pd

# Create a DataFrame from the stacked values
df_result = pd.DataFrame(stacked_values, columns=["Original y", "Estimated y"])
df_result["Error Percentage"] = df_result.apply(
    lambda row: abs(row["Original y"] - row["Estimated y"]) / row["Original y"] * 100,
    axis=1,
)

# Save the DataFrame to a CSV file
df_result.to_csv("DTSM.csv", index=False)
```
